```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FormatStrFormatter
import matplotlib.font_manager as fm
import palettable
plot_root = '../../plots/renewable'
```

```python
df1 = pd.read_csv('../../output_renewable/ces0_89_86_higheps.csv')
df2 = pd.read_csv('../../output_sig0/ces0.csv')
df3 = pd.read_csv('../../output_renewable/ces0_89_86.csv')
df4 = pd.read_csv('../../output_renewable/ces0_80_86.csv')
```

```python
df1 = df1[df1['region_scenario'] == 3]
df2 = df2[(df2['region_scenario'] == 3) & (df2['tax_sce']=='Unilateral')]
df3
```

```python
plt.plot(df1['varphi'],df1['Ceystar_prime'],linewidth=1, label = 'Ceystar_prime')
plt.plot(df1['varphi'],df1['Cey_prime'],linewidth=1, label = 'Cey_prime')
plt.plot(df1['varphi'],df1['Cex_prime'],linewidth=1, label = 'Cex_prime')
plt.plot(df1['varphi'],df1['Cem_prime'],linewidth=1, label = 'Cem_prime')
plt.legend()
```

```python
plt.plot(df1['varphi'],df1['Qe1_prime'],linewidth=1, label = 'Qe1')
plt.plot(df1['varphi'],df1['Qe2_prime'],linewidth=1, label = 'Qe2_prime')
plt.plot(df1['varphi'],df1['Qe1star_prime'],linewidth=1, label = 'Qe1star')
plt.plot(df1['varphi'],df1['Qe2star_prime'],linewidth=1, label = 'Qe2star_prime')
plt.legend()
```

```python
plt.plot(df1['varphi'],df1['jxbar_prime'],linewidth=1, label = 'jxbar_prime')
plt.plot(df1['varphi'],df1['j0_prime'],linewidth=1, label = 'j0_prime')
plt.legend()
```

```python
df_plot = df1
plt.plot(df_plot['varphi'],df_plot['Ge_prime'],linewidth=1, label = 'Ge_prime')
plt.plot(df_plot['varphi'],df_plot['Gestar_prime'],linewidth=1, label = 'Gestar_prime')
plt.plot(df_plot['varphi'],df_plot['Qe2_prime'],linewidth=1, label = 'Qe2_prime')
plt.plot(df_plot['varphi'],df_plot['Qe2star_prime'],linewidth=1, label = 'Qe2star_prime')
plt.legend()

plt.xlabel('varphi')
plt.ylabel('Energy')
plt.savefig(plot_root + '/varphi_ge_higheps.pdf',format='pdf', bbox_inches = 'tight')
```

```python
df_plot = df4
plt.plot(df_plot['varphi'],df_plot['Ge_prime'],linewidth=1, label = 'Ge_prime')
plt.plot(df_plot['varphi'],df_plot['Gestar_prime'],linewidth=1, label = 'Gestar_prime')
plt.plot(df_plot['varphi'],df_plot['Qe2_prime'],linewidth=1, label = 'Qe2_prime')
plt.plot(df_plot['varphi'],df_plot['Qe2star_prime'],linewidth=1, label = 'Qe2star_prime')
plt.legend()

plt.xlabel('varphi')
plt.ylabel('Energy')
plt.savefig(plot_root + '/varphi_ge.pdf',format='pdf', bbox_inches = 'tight')
```

```python
plt.xlim(-10,0)
x = 'welfare_noexternality'
y = 'Qeworld_chg'
df_plot=df1[df1['varphi'] <= 4.5]
df_plot['Qeworld_chg']=-(df_plot['Qe1_prime'] + df_plot['Qe1star_prime'] -df_plot['Qe1_prime'].iat[0]-df_plot['Qe1star_prime'].iat[0]) / (df_plot['Qe1_prime'].iat[0]+df_plot['Qe1star_prime'].iat[0]) * 100
plt.plot(df_plot[x],df_plot[y],linestyle='solid', label = 'Higher elasticity of renewable energy')
dot=df_plot[df_plot['varphi']==2] #puretp
plt.plot(dot[x],dot[y], marker='x', markersize=3, color='red') 

df_plot=df2
df_plot['Qeworld_chg']=-(df_plot['Qeworld_prime']-df_plot['Qeworld_prime'].iat[0]) / df_plot['Qeworld_prime'].iat[0] * 100
plt.plot(df_plot[x],df_plot[y],linestyle='solid', label = 'No renewable energy')
dot=df_plot[df_plot['varphi']==2] #puretp
plt.plot(dot[x],dot[y], marker='x', markersize=3, color='red') 

df_plot=df3
df_plot['Qeworld_chg']=-(df_plot['Qe1_prime'] + df_plot['Qe1star_prime'] -df_plot['Qe1_prime'].iat[0]-df_plot['Qe1star_prime'].iat[0]) / (df_plot['Qe1_prime'].iat[0]+df_plot['Qe1star_prime'].iat[0]) * 100
plt.plot(df_plot[x],df_plot[y],linestyle='solid', label = 'With renewable energy')
dot=df_plot[df_plot['varphi']==2] #puretp
plt.plot(dot[x],dot[y], marker='x', markersize=3, color='red') 

df_plot=df4[df4['varphi'] < 9]
df_plot['Qeworld_chg']=-(df_plot['Qe1_prime'] + df_plot['Qe1star_prime'] -df_plot['Qe1_prime'].iat[0]-df_plot['Qe1star_prime'].iat[0]) / (df_plot['Qe1_prime'].iat[0]+df_plot['Qe1star_prime'].iat[0]) * 100
plt.plot(df_plot[x],df_plot[y],linestyle='solid', label = 'Higher initial share of renewable')
dot=df_plot[df_plot['varphi']==2] #puretp
plt.plot(dot[x],dot[y], marker='x', markersize=3, color='red') 

plt.legend()
ylab = 'Global emissions reductions (% of BAU)'
xlab = 'Change in total consumption \n(% of initial goods consumption)'
plt.xlabel(xlab)
plt.ylabel(ylab)

plt.savefig(plot_root + '/fig4.pdf', format='pdf', bbox_inches = 'tight')
```
